### Importing The Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm # Support Vector Maching
from sklearn.metrics import accuracy_score
```

```python
import warnings

# Suppress specific warning
warnings.filterwarnings("ignore", category=UserWarning)
```

### Data Collection And Processing

```python
# Importing the data to a pandas dataframe
diabetes_dataset = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Datasets/diabetes.csv')
```

```python
# Printing the first 5 rows
diabetes_dataset.head()
```

```python
# Printing the dimension of the dataset
diabetes_dataset.shape
```

```python
# Statistical measures of the dataset
diabetes_dataset.describe()
```

### Meaning of the outcome  
* 0 --> Non-diabetic  
* 1 --> Diabetic

```python
# Printing the number of lables
diabetes_dataset['Outcome'].value_counts()
```

```python
# Mean of each label in the outcome column
diabetes_dataset.groupby('Outcome').mean()
```

```python
# Seprating the datas and labels
x = diabetes_dataset.drop(columns = 'Outcome', axis = 1)
y = diabetes_dataset['Outcome']
```

```python
x.head()
```

```python
y.head()
```

### Data Standardization

```python
scaler = StandardScaler()
```

```python
scaler.fit(x)
std_x = scaler.transform(x)
x = std_x
```

```python
# Standarized data s
x.std()
```

### Spliting the data to training and testing data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 2, stratify = y)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

### Model Training

```python
classifier = svm.SVC(kernel = 'linear')
```

```python
# Training the svm with the training data
classifier.fit(x_train, y_train) # classifier.fit(data, label)
```

### Model Evaluation

```python
# Accuracy score on the training data
x_train_prediction = classifier.predict(x_train)
x_train_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print(x_train_accuracy)
```

```python
# Accuracy score on the testing data
x_test_prediction = classifier.predict(x_test)
x_test_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print(x_test_accuracy)
```

### Predictive System

```python
input_data = (5,166,72,19,175,25.8,0.587,51)

# changing the input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# standardize the input data
std_data = scaler.transform(input_data_reshaped)
# print(std_data)

prediction = classifier.predict(std_data)
# print(prediction)

if (prediction[0] == 0):
  print('The Person Is Not Diabetic')
else:
  print('The Person Is Diabetic')
```
